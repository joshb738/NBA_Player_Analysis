```python
# Renamed check_data.csv to edited_data after taking out the duplicat names
# You can start the notebook from here if you do not want to take out the duplicates yourself
edited_df = pd.read_csv('Data/edited_data2.csv')
edited_df.head()
```

```python
a_df = edited_df.drop(columns = ['GS', 'Pos', 'position', 'year_end', 'college', 'Unnamed: 0', 'birth_date', 'name',])
a_df.head()
```

```python
b_df = pd.DataFrame(columns = a_df.columns)
b_df.head()
```

```python
for index, row in a_df.iterrows():
    subset = a_df[(a_df["Year"]==row["Year"]) & (a_df["Player"]==row["Player"]) & (a_df["year_born"]==row["year_born"])]
    if len(subset) == 1:
        b_df = b_df.append(subset)
    elif subset[subset["Tm"]=="TOT"].index[0] not in b_df.index:
        b_df = b_df.append(subset[subset["Tm"]=="TOT"])
```

```python
b_df
```

```python
b_df["Status"] = (b_df["Year"] - b_df["year_start"] +1)
b_df
```

```python
b_df = b_df.drop(columns = ['Tm', 'height', 'blanl', 'blank2','FG%','3P%', '2P%', 'FT%'], axis=1)
b_df.head()
```

```python
b_df = b_df[b_df["G"] > 10]
b_df
```

```python
clustered_df = pd.read_csv ('clustered_dataset.csv')
```

```python
cols = ["Player","year_born"]
for i in range(1,22):
    cols.extend([f"Cluster {i}",f"VORP {i}",f"WS {i}"])
c_df = pd.DataFrame(columns = cols)
c_df.head()
```

```python
for index, row in clustered_df.iterrows():
    if len(b_df[(b_df["Player"]==row["Player"]) & (b_df["year_born"]==row["year_born"])])>0:
        subset = b_df[(b_df["Player"]==row["Player"]) & (b_df["year_born"]==row["year_born"])]
        if (subset["Player"].iloc[0],subset["year_born"].iloc[0]) not in zip(c_df["Player"],c_df["year_born"]):
            new_row = {}
            new_row["Player"] = subset["Player"].iloc[0]
            new_row["year_born"] = subset["year_born"].iloc[0]
            for index2, sub in subset.iterrows():
                new_row[f"VORP {sub['Status']}"] = sub["VORP"]
                new_row[f"WS {sub['Status']}"] = sub["WS"]
            c_df = c_df.append(new_row,ignore_index=True)
```

```python
c_df
```

```python
clustered_df
```

```python
clustered_df2 = clustered_df.copy()
for index,row in clustered_df.iterrows():
    clustered_df2.iloc[index] = clustered_df.iloc[index].fillna(c_df.iloc[index])
```

```python
clustered_df2
```

```python
clustered_df2.to_csv('for_vis_dataset.csv', index=False)
```
