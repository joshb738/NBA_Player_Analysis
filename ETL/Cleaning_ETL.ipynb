```python
# Renamed check_data.csv to edited_data after taking out the duplicat names
# You can start the notebook from here if you do not want to take out the duplicates yourself
edited_df = pd.read_excel('Data/edited_data.xlsx')
edited_df.head()
```

```python
a_df = edited_df.drop(columns = ['GS', 'Pos', 'position', 'year_end', 'college', 'Unnamed: 0', 'birth_date', 'name',])
a_df.head()
```

```python
len(a_df)
```

```python
b_df = pd.DataFrame(columns = a_df.columns)
b_df.head()
```

```python
for index, row in a_df.iterrows():
    subset = a_df[(a_df["Year"]==row["Year"]) & (a_df["Player"]==row["Player"]) & (a_df["year_born"]==row["year_born"])]
    if len(subset) == 1:
        b_df = b_df.append(subset)
    elif subset[subset["Tm"]=="TOT"].index[0] not in b_df.index:
        b_df = b_df.append(subset[subset["Tm"]=="TOT"])
```

```python
b_df
```

```python
b_df["Status"] = (b_df["Year"] - b_df["year_start"] +1)
b_df
```

```python
# Seperate feet and inches and convert all measurments to inches
b_df['feet'] = b_df.height.str.split('-').str[0].astype(np.float) * 12
b_df['inch'] = b_df.height.str.split('-').str[1].astype(np.float)

# Combine measurements in new column "height(inch)"
b_df["height(inch)"] = b_df.feet+b_df.inch
```

```python
b_df = b_df.drop(columns = ['Tm', 'height', 'feet', 'inch', 'blanl', 'blank2','FG%','3P%', '2P%', 'FT%'], axis=1)
b_df.head()
```

```python
b_df = b_df[b_df["G"] > 10]
b_df
```
